PTRATIO: PUPIL-TEACHER RATIO IN TOWN

LSTAT:   %LOWER STATUS OF POPULATION

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(strategy='median')
imputer.fit(housing)
```

```python
imputer.statistics_
```

```python
x=imputer.transform(housing)
```

```python
housing_tr=pd.DataFrame(x,columns=housing.columns)
```

```python
print(housing_tr)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
my_pipeline=Pipeline([
    ('imputer',SimpleImputer(strategy='median')),
    ('std_scaler',StandardScaler()),

])
```

```python
housing_num=my_pipeline.fit_transform(housing)
```

```python
housing_num
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
model=LinearRegression()
model.fit(housing_num,housing['MEDV'])

```

```python
train_data=housing.iloc[:5]
```

```python
data=my_pipeline.transform(train_data)
```

```python
model.predict(data)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
housing_predictions=model.predict(housing_num)
```

```python
mse=mean_squared_error(housing['MEDV'],housing_predictions)
rmse=np.sqrt(mse)
```

```python
rmse
```
